Lab 3 - Conditional Least Square Adjustment
Given a network of stations, we are provided the change in height measurements between each station where an observation was completed. Using C++ and Eigen we will adjust the stations to satisfy conditional equations.
Python will be used for pre-processing of the data into an easier to read format, verify linear independence of the equations, and graphical interpretation of residuals.

```python
import pandas as pd
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

As the B matriz was populated in Excel, pandas will be used to import the matrix of partial derivatives, verify the rank and write it to .txt.

```python
B = pd.read_excel('B_matrix.xlsx', usecols='B:U', header=0 )
bRank = LA.matrix_rank(B)
print(bRank)
B.to_csv('bMatrix.txt', sep='\t', header=False, index=False)
print("bMatrix.txt created")
```

The same process will be used for J

```python
J = pd.read_excel('J_matrix.xlsx', usecols='B:U', header=0 )
jRank = LA.matrix_rank(J)
print(jRank)
J.to_csv('jMatrix.txt', sep='\t', header=False, index=False)
print("jMatrix.txt created")
```

```python
residuals = pd.read_table('v_hat_Lab_3.txt', header=None)
x = range(len(residuals))

plt.figure(figsize=(8,6))
plt.plot(x, residuals, label='Residuals')
plt.axhline(0, linestyle='--')
plt.xlabel("Height Measurement"); plt.ylabel("Residual"); plt.title("Residuals vs Station Measurements")
plt.grid(True, alpha=.3); plt.legend();
```
